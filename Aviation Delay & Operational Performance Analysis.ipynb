# Data Cleaning and Feature Selection

The previous analysis revealed that a significant portion of the dataset contains features with extremely high missing values (often above 98–100%). These columns do not contribute meaningful information for analysis or prediction and must be removed.

In real-world aviation analytics, feature reduction is a critical step because:

- Legacy flight systems generate sparse telemetry fields
- Not all recorded variables are operationally relevant
- High-dimensional sparse data reduces model performance and interpretability

In this step, we will:
    
- Remove columns with excessive missing values
- Retain only operationally meaningful features
- Prepare a clean dataset for exploratory analysis and machine learning


````python
# ==========================================
# Data Cleaning - Feature Reduction
# ==========================================

# 1. Remove columns with >95% missing values
threshold = 0.95 * len(df)
df_clean = df.loc[:, df.isnull().sum() < threshold]

print("Original Shape:", df.shape)
print("Cleaned Shape:", df_clean.shape)

# 2. Drop explicitly useless placeholder columns if still present
drop_cols = [col for col in df_clean.columns if "Unused" in col or "Feature_" in col]

df_clean = df_clean.drop(columns=drop_cols, errors='ignore')

print("\nAfter Removing Sparse Features:")
print("Shape:", df_clean.shape)

# 3. Show remaining columns
print("\nRemaining Columns:")
print(df_clean.columns.tolist())

# 4. Rename working dataset
df = df_clean.copy()

print("\nData cleaning completed successfully.")
````

## Train-Test Split for Aviation Delay Prediction

Before training any machine learning model, it is essential to separate the dataset into training and testing subsets.

This ensures that model performance is evaluated using unseen data, providing a realistic estimate of how the model will perform in real-world aviation operations.

---

## Key Principles in This Step

In aviation predictive modeling, preventing **data leakage** is critical.

Data leakage occurs when information unavailable at prediction time is accidentally included during model training, resulting in unrealistically high performance.

To avoid this, we will:

1. Encode categorical variables into numeric format
2. Separate features (`X`) from the target variable (`y`)
3. Create independent training and testing datasets
4. Ensure reproducibility using a fixed random state

---

## Why This Matters in Aviation Analytics

Proper train-test separation ensures:

- Reliable model evaluation
- Better generalization across airports and airlines
- Realistic delay forecasting performance
- Reduced risk of overfitting

---



````python

# ==========================================
# Train-Test Split (Aviation Delay Model)
# ==========================================

from sklearn.model_selection import train_test_split

# Create working copy
data = model_df.copy()

# ------------------------------------------
# Encode categorical variables
# ------------------------------------------

categorical_cols = [
    "Airline",
    "Operating_Airline",
    "Origin",
    "Destination"
]

data_encoded = pd.get_dummies(
    data,
    columns=categorical_cols,
    drop_first=True
)

# ------------------------------------------
# Define features and target
# ------------------------------------------

X = data_encoded.drop(
    "Arrival_Delay",
    axis=1
)

y = data_encoded["Arrival_Delay"]

# ------------------------------------------
# Train-Test Split
# ------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

# ------------------------------------------
# Output Shapes
# ------------------------------------------

print("Train-Test Split Completed Successfully!")
print("=" * 50)

print("\nTraining Set:")
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)

print("\nTesting Set:")
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)


````

````python
# ==========================================
# Random Forest Regressor - Baseline Model
# ==========================================

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# ------------------------------------------
# Initialize model
# ------------------------------------------

rf_model = RandomForestRegressor(
   n_estimators=100,
   random_state=42,
   n_jobs=-1
)

# ------------------------------------------
# Train model
# ------------------------------------------

print("Training Random Forest model...")
rf_model.fit(X_train, y_train)

# ------------------------------------------
# Predictions
# ------------------------------------------

print("Generating predictions...")
y_pred = rf_model.predict(X_test)

print("Predictions completed.")
print("Number of predictions:", len(y_pred))

# ------------------------------------------
# Evaluation Metrics
# ------------------------------------------

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("\nModel Performance")
print("-" * 40)
print(f"MAE : {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²  : {r2:.4f}")
````

At this stage, we transition from machine learning evaluation into business intelligence interpretation.

- What actionable insights airlines can use to reduce delays

Objective of This Step

We will:

- Classify flights into risk categories based on delay magnitude
- Analyze how model error varies across risk groups
- Summarize operational impact distribution
- Translate findings into airline and airport recommendations

## Why This Matters in Aviation Analytics

Airlines do not operate on raw predictions — they operate on risk levels:

- Low-risk flights → normal operations
- Medium-risk flights → monitoring required
- High-risk flights → operational intervention needed

This step bridges machine learning and real-world aviation decision systems.


````python
# ==========================================
# Aviation Risk Classification
# ==========================================

# Create evaluation dataframe
results_df = pd.DataFrame({
   "Actual": y_test,
   "Predicted": y_pred
})

# ------------------------------------------
# Define risk categories based on actual delay
# ------------------------------------------

def classify_risk(delay):
   if delay <= 10:
       return "HEALTHY"
   elif delay <= 30:
       return "WARNING"
   else:
       return "CRITICAL"

results_df["Risk_Level"] = results_df["Actual"].apply(classify_risk)

# ------------------------------------------
# Risk distribution
# ------------------------------------------

risk_counts = results_df["Risk_Level"].value_counts()

print("Engine Risk Distribution")
print("-" * 40)
print(risk_counts)

# ------------------------------------------
# Error by risk level
# ------------------------------------------

results_df["Abs_Error"] = np.abs(results_df["Actual"] - results_df["Predicted"])

error_by_risk = results_df.groupby("Risk_Level")["Abs_Error"].mean()

print("\nAverage Error by Risk Level")
print("-" * 40)
print(error_by_risk)
````
